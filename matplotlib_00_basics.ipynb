```python
# A figure object is the outer container, 
# which can have multiple axes objects
# The axes is not the plural of axis,
# it is what we think of as a plot or a graph
# (think of it as 'ax'es)

fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
ax1.scatter(3,4,s=30)
ax1.text(3+0.3,4+0.3,'p1',size=10)
ax1.set_xticks(range(-5,6)) # -5 ~ 5 x-axis
ax1.set_yticks(range(-5,6))
ax1.grid()
ax1.set_axisbelow(True) # grid is below the point
ax1.set_aspect('equal',adjustable='box')
ax1.spines['left'].set_position('zero')
ax1.spines['bottom'].set_position('zero')
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(4,3,s=30)
ax2.text(4+0.3,3+0.3,'p2',size=10)
ax2.set_xticks(range(-5,6))
ax2.set_yticks(range(-5,6))
ax2.grid()
# ax2.set_axisbelow(True)
ax2.set_aspect('equal',adjustable='box')
ax2.spines['left'].set_position('zero')
ax2.spines['bottom'].set_position('zero')
ax2.spines['right'].set_color('silver')
ax2.spines['top'].set_color('none')

plt.show()
```

```python
# quiver draws an arrow
# representing an vector

start = np.array([0,0])
end = np.array([3,2])

fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
ax1.quiver(start[0],start[1],end[0],end[1],angles='xy',scale_units='xy',scale=1)
# angles='xy' # Arrows point from (x,y) to (x+u,x+v)
# To plot vectors in the x-y plane, with u and v having the same units as x and y,
# you should use angles='xy', scale_units='xy', scale=1

ax1.text(end[0],end[1],'a',size=20)

ax1.set_xticks(range(0,6))
ax1.set_yticks(range(0,6))
ax1.grid()
ax1.set_axisbelow(True)
ax1.set_aspect('equal', adjustable='box')

ax1.spines['left'].set_position('zero')
ax1.spines['bottom'].set_position('zero')


ax2 = fig.add_subplot(1,2,2)
ax2.quiver(start[0],start[1],end[0],end[1],color='red',angles='xy',scale_units='xy',scale=1)
ax2.text(end[0],end[1],'a',size=20)

ax2.set_xticks(range(0,6))
ax2.set_yticks(range(0,6))
ax2.grid()
ax2.set_axisbelow(True)
ax2.set_aspect('equal', adjustable='box')

ax2.spines['left'].set_position('zero')
ax2.spines['bottom'].set_position('zero')
ax2.spines['top'].set_color('silver')
ax2.spines['right'].set_color('silver')

plt.show()
```

```python
# subplot's'

A = np.array([2,-2]) # direction components of a vector, not xy coordinates
B = np.array([3,4])
C = np.array([5,2])

fig, ax = plt.subplots(1,3) # one row, three columns

ax[0].title.set_text('A')
ax[1].title.set_text('B')
ax[2].title.set_text('C')

ax[0].quiver(0,0,A[0],A[1],color='blue',angles='xy',scale_units='xy',scale=1)

ax[1].quiver(0,0,A[0],A[1],color='blue',angles='xy',scale_units='xy',scale=1)
ax[1].quiver(2,-2,B[0],B[1],color='green',angles='xy',scale_units='xy',scale=1)

ax[2].quiver(0,0,A[0],A[1],color='blue',angles='xy',scale_units='xy',scale=1)
ax[2].quiver(2,-2,B[0],B[1],color='green',angles='xy',scale_units='xy',scale=1)
ax[2].quiver(0,0,C[0],C[1],color='purple',angles='xy',scale_units='xy',scale=1)

for i in range(3):
    ax[i].axis([-2,3,-4,5]) # axis([xmin,xmax,ymin,ymax])
    ax[i].set_xticks(range(-2,6))
    ax[i].set_yticks(range(-4,6))
    ax[i].grid(True)
    ax[i].set_axisbelow(True)
    ax[i].set_aspect('equal', adjustable='box')
    ax[i].spines['left'].set_position('zero')
    ax[i].spines['bottom'].set_position('zero')
    ax[i].spines['right'].set_color('none')
    ax[i].spines['top'].set_color('none')

plt.show()
```

```python
# A contour plot is representing a 3D surface 
# by plotting constant z slices(contours)

# For 2D density and contour plots,
# see https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html

def cos_d_from_o(x,y):
    d_from_o = np.sqrt(x**2 + y**2) # np.sqrt((x-0)**2 + (y-0)**2)
    return np.cos(d_from_o)

x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)

X,Y = np.meshgrid(x,y) # x: from left to right, y: from top to bottom
Z = cos_d_from_o(X,Y)


fig = plt.figure()
ax = plt.axes(projection='3d')
p = ax.contour3D(X,Y,Z, 100, cmap='ocean') # contour3D (X,Y),Z

fig.colorbar(p, ax=ax) # colorbar for 3D

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(-6,7)
ax.set_ylim(-6,7)
ax.set_zlim(-2,3)

ax.view_init(elev=30, azim=30) # view angle, elevation and azimuthal angles

plt.show()
```

```python
# plot_wireframe (3D plot)

def sin_d_from_o(x,y):
    d_from_o = np.sqrt(x**2 + y**2) # np.sqrt((x-0)**2 + (y-0)**2)
    return np.sin(d_from_o)

x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)

X,Y = np.meshgrid(x,y) # x: from left to right, y: from top to bottom
Z = sin_d_from_o(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(X,Y,Z, color='black')
ax.set_title('Wireframe')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(-6,7)
ax.set_ylim(-6,7)
ax.set_zlim(-2,3)

ax.view_init(elev=50, azim=30) # view angle, elevation and azimuthal angles

plt.show()
```

```python
# plot_surface
# (polygon)

ax = plt.axes(projection='3d')
p = ax.plot_surface(X,Y,Z, rstride=2, cstride=2, cmap='viridis_r', edgecolor='none', alpha=0.8)

ax.set_title('Surface')
fig.colorbar(p, ax=ax)


plt.show()
```

```python
# partial polar grid

def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))

r = np.linspace(0,7,20)
theta = np.linspace(-np.pi, 0.5*np.pi, 40) # angle for making slice

r, theta = np.meshgrid(r, theta) # imagine arcs with each radius length

X = r*np.sin(theta) # disassemble arc to its x
Y = r*np.cos(theta) # disassemble arc to its y
Z = f(X,Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap='ocean', edgecolor='black')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
# 3D scatter and plot_trisurf

def cos_d_from_o(x,y):
    return np.cos(np.sqrt(x**2+y**2))

# np.random.random returns random floats in interval [0.0,1.0)
theta = 2 * np.pi * np.random.random(700)
radius = 5 * np.random.random(700)
x = radius*np.sin(theta)
y = radius*np.cos(theta)
# You can use np.ravel() to make contiguous flattened array(1dim)
# if necessary
z = cos_d_from_o(x,y)

# figure and axes1, axes2
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1,projection='3d') # 3d, not 3D
ax2 = fig.add_subplot(1,2,2,projection='3d')

# 3D scatter
p1 = ax1.scatter(x, y, z, c=z, cmap='summer', linewidth=0.1, edgecolor='black')

# plot_trisurf
# It draws triangles from adjacent points
# (it's why we need contiguous flattened array(1dim array) for x,y,z)
x = np.ravel(x) # contiguous flattened array(1dim)
y = np.ravel(y)
z = cos_d_from_o(x, y)
p2 = ax2.plot_trisurf(x,y,z, cmap='winter', edgecolor='none')

plt.show()
```

```python
# Visualizing Mobius strip
# Think about parametrization

theta = np.linspace(0, 2*np.pi, 50) # position of the loop about its center
width = np.linspace(-0.5, 0.5, 5)
width, theta = np.meshgrid(width,theta)

# twisting 180 degrees of the strip about its axis, 
# along with the change of theta
# (phi rotates the half, while theta rotates one round)
phi = theta/2 # changing it to theta/3 will be insufficient to make Mobius strip

rd = 1 + width*np.cos(phi) # loop radius + distance made from the strip

x = np.ravel(rd*np.cos(theta)) # disassemble rd to find the x coordinate
y = np.ravel(rd*np.sin(theta)) # disassemble rd to find the y coordinate

z = np.ravel(width*np.sin(phi)) # the strip slope


# Triangulate parameter space to determine the triangles
from matplotlib.tri import Triangulation
triangles = Triangulation(np.ravel(theta), np.ravel(width)).triangles

ax = plt.axes(projection='3d')
ax.plot_trisurf(x,y,z, triangles=triangles, cmap='summer_r', linewidths=0.2)

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(elev=40,azim=40)

plt.show()
```
